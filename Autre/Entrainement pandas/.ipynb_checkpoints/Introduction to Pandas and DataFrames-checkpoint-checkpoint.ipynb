# Les dataframes

## Initisaliser un dataframe vide

```python
df_empty = pd.DataFrame()
print(df_empty)
```

```python
noms = ["Adrien","Brice","Helena","Enzo","Kim"]
df = pd.DataFrame(noms)
df
```

```python
Naissance = [2001,2002,2003,2003,2000]
list_of_tuples = list(zip(noms, Naissance))
 
list_of_tuples
```

```python
df_tuples = pd.DataFrame(list_of_tuples)
df_tuples
```

```python
df_tuples = pd.DataFrame(list_of_tuples,columns=["Nom","Naissance"],index = ["1er","2eme","3eme","4eme","5eme"])
df_tuples
```

## Selcetionner des parties d'un dataframe

### Selectionner une colonne

```python
df["Rang"]
```

```python
df[["Rang","Mineure"]]
```

### Selectionner les n premières ou dernières lignes

```python
df.head(2)
```

```python
df.tail(3)
```

### Selectionner une ligne à partir de son index

```python
df.loc[2]
```

```python
df.loc[3:5]
```

### Selectionner des lignes et des colonnes spécifiques

```python
df.loc[2:5,"Mineure":'Année de naissance']
```

### Accerder aux colonnes à partir de leur numéro plutôt que leur nom

```python
df.iloc[3:6,1:3]
```

```python
df[df["Année de naissance"]<2003 & ((df["Favori"]=="Analyse") | (df["Nom eleve"]=="Brice"))]
```
