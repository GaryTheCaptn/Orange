# Résultats des modélisations
Dans ce fichier nous allons récapituler et faire des statistiques sur les résultats des modèles créés. L'objectif est de déterminer quels sont les modèles les plus pertinents mais aussi les risques qui leur sont associés.

```python
 #Quelques importations
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Données sous forme de liste de [Spécificité,VPN]
arbre_decision = pd.DataFrame({"Type":["arbre_decision" for x in range (0,10)],"Specificite": [0.17,0.15,0.26,0.19,0.19,0.17,0.34,0.26,0.33,0.27],"VPN":[0.94,0.93,0.96,0.76, 0.76,0.75,0.87,0.92,0.88,0.81]})
random_forest =  pd.DataFrame({"Type":["random_forest" for x in range (0,10)],"Specificite": [0.25,0.28,0.24,0.23,0.25,0.23,0.21,0.2,0.16,0.19],"VPN":[0.95,0.96,1,0.95,1,0.91,1,0.91,1,0.95]})
neighbors =  pd.DataFrame({"Type":["neighbors" for x in range (0,10)],"Specificite": [0.3,0.304,0.371,0.385,0.382,0.356,0.367,0.367,0.32,0.36],"VPN":[0.97,0.971,0.973,0.918,0.929,0.923,0.9,0.9,0.94,0.97]})
regression= pd.DataFrame({"Type":["regression" for x in range (0,5)],"Specificite": [0.39,0.42,0.36,0.4,0.36],"VPN":[0.73,0.61,0.62,0.55,0.59]})
SVM =  pd.DataFrame({"Type":["svm" for x in range (0,5)],"Specificite": [0.45,0.49,0.51,0.54,0.55],"VPN":[0.64,0.67,0.64,0.66,0.64]})
naive_bayes = pd.DataFrame({"Type":["naive_bayes" for x in range (0,5)],"Specificite": [0.82,0.78,0.76,0.8,0.79],"VPN":[0.46,0.46,0.43,0.42,0.53]})

```

```python
df = pd.concat([arbre_decision,random_forest,neighbors,regression,SVM,naive_bayes])
df.reset_index(drop=True)
```

```python
sns.lmplot( x="Specificite", y="VPN", data=df, fit_reg=False, hue='Type', legend=False)
plt.legend(loc='upper right')
plt.show()
```

On recherche les modèles donnant des résultats qui ont la plus grand valeur prédictive négative et si possible une bonne spécificité. Les modèles de régression, svm et naive bayes ne sont pas adaptés pour notre cas mais on peut retenir les modèles d'arbre de décision, de forêts aléatoires et de k plus proches voisins.

La méthode des arbres de décisions a cependant le défaut d'avoir des résultats dépendants beaucoup du dataset d'entrainement mais il se pourrait que les 2 autres types de modèles soient en sur-apprentissage et ne soient pas du tout utilisable dans la réalité.
